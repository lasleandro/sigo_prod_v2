```python
mycursor.execute("SHOW TABLES")
tables_list = mycursor.fetchall()
tables_list = [item[0] for item in tables_list]
tables_list

if 'screening_table' not in tables_list:
    table = """
         CREATE TABLE IF NOT EXISTS screening_table(
         main_id int(11) NOT NULL AUTO_INCREMENT,
         mail_id varchar(50) NOT NULL,
         thread_id varchar(50) NOT NULL,
         mail_date_received date NOT NULL,
         mail_hms_received time NOT NULL,
         mail_subject varchar(10000) NOT NULL,
         mail_from varchar(60) NOT NULL,
         mail_body longblob NULL,
         mail_body_snippet varchar(5000) NULL,
         has_attachment enum('0', '1') NOT NULL,
         attachment_folder_link varchar(50),
         case_id varchar(50),
         external_ref_number varchar(50),
         date_created timestamp NOT NULL,
         date_updated timestamp default current_timestamp on update current_timestamp,
         PRIMARY KEY (main_id))
    """

    mycursor.execute(table)

    mydb.commit()
    print('Table created!')
    
else:
    print('Table already exists!')
```

```python
data_path = os.path.join(os.path.dirname(cwd), '_collect_control')
print(data_path)
```

```python
df = pd.read_csv(os.path.join(data_path, 'mail_data.csv'), sep = '\t', encoding = 'utf-8')
df = df.drop('mail_date_received_full', axis = 1)
df = df.tail(100)
```

```python
df['mail_date_received'] = pd.to_datetime(df['mail_date_received'])
df['mail_hms_received'] = pd.to_datetime(df['mail_hms_received']).dt.time
```

```python
df['mail_body'] = df['mail_body'].str.encode('utf-8')
```

```python
df['has_attachment'] = df['has_attachment'].astype(str)
```

```python
# Insert whole DataFrame into MySQL
df.to_sql('screening_table', con = engine, if_exists = 'append', index = False, chunksize = 1000)
```
